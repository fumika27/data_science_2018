```python
%matplotlib inline
import sys 
sys.path.append('..')
from preamble import *
```

```python
# 데이터셋을 만듦
X, y = mglearn.datasets.make_forge()

#산점도를 그림
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["class 0", "class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
```

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("feature:")
plt.ylabel("target")
```

```python
# blobs와 wave의 차이를 알아보기 위해 수행해봄
X, y = mglearn.datasets.make_blobs(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("feature:")
plt.ylabel("target")
```

* 특성이 적은 데이터셋 (저차원 데이터셋)에서 얻은 직관이 특성이 많은 데이터셋 (고차원 데이터셋)에서 그대로 유지되지 않을 수 있음

- scikit-learn에는 유방암 종양의 임상 데이터를 기록해놓은 위스콘신 유방암 데이터 셋이 있음
    - 각 종양은 양성 (해롭지 않은 종양)과 악성 (암 종양)으로 레이블링되어 있음
    - 조직 데이터를 기반으로 종양이 양성인지 악성인지를 예측할 수 있도록 학습

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print("cancer.keys(): \n{}".format(cancer.keys()))
```

```python
print("feature of cancer: {}".format(cancer.data.shape))
```

    - 569개 데이터 중 212개는 악성이고 357개는 양성

```python
print("the number ofr samples per class:\n{}".format({n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))
```

```python
print("feature name:\n{}".format(cancer.feature_names))
```
